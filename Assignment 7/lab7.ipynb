```python
# ==========================================
# IMPORT LIBRARIES
# ==========================================
import nltk
import re
import math
import pandas as pd

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# ==========================================
# DOWNLOAD REQUIRED DATA (RUN ONCE)
# ==========================================
nltk.download('punkt')
nltk.download('punkt_tab')   # important fix
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# ==========================================
# INPUT TEXT
# ==========================================
text = "Tokenization is the first step in text analytics. It breaks text into smaller parts."

# ==========================================
# STEP 1: SENTENCE TOKENIZATION
# ==========================================
print("\n--- Sentence Tokenization ---")
sent_tokens = sent_tokenize(text)
print(sent_tokens)

# ==========================================
# STEP 2: WORD TOKENIZATION
# ==========================================
print("\n--- Word Tokenization ---")
word_tokens = word_tokenize(text)
print(word_tokens)

# ==========================================
# STEP 3: STOPWORD REMOVAL
# ==========================================
print("\n--- Stopword Removal ---")
stop_words = set(stopwords.words("english"))

clean_text = re.sub('[^a-zA-Z]', ' ', text)
tokens = word_tokenize(clean_text.lower())

filtered_words = []
for w in tokens:
    if w not in stop_words:
        filtered_words.append(w)

print(filtered_words)

# ==========================================
# STEP 4: STEMMING
# ==========================================
print("\n--- Stemming ---")
ps = PorterStemmer()

words = ["wait", "waiting", "waited", "waits"]

for w in words:
    print(w, "->", ps.stem(w))

# ==========================================
# STEP 5: LEMMATIZATION
# ==========================================
print("\n--- Lemmatization ---")
lemmatizer = WordNetLemmatizer()

text2 = "studies studying cries cry"
tokens2 = word_tokenize(text2)

for w in tokens2:
    print(w, "->", lemmatizer.lemmatize(w))

# ==========================================
# STEP 6: POS TAGGING
# ==========================================
print("\n--- POS Tagging ---")
data = "The pink sweater fits her perfectly"
words = word_tokenize(data)

for w in words:
    print(nltk.pos_tag([w]))

# ==========================================
# STEP 7: TF-IDF (MANUAL)
# ==========================================
print("\n--- TF-IDF Manual Calculation ---")

documentA = "Jupiter is the largest planet"
documentB = "Mars is the fourth planet from the sun"

# Tokenize
bagOfWordsA = documentA.lower().split()
bagOfWordsB = documentB.lower().split()

# Unique words
uniqueWords = set(bagOfWordsA).union(set(bagOfWordsB))

# Word count
numOfWordsA = dict.fromkeys(uniqueWords, 0)
numOfWordsB = dict.fromkeys(uniqueWords, 0)

for word in bagOfWordsA:
    numOfWordsA[word] += 1

for word in bagOfWordsB:
    numOfWordsB[word] += 1

# TF
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    totalWords = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(totalWords)
    return tfDict

tfA = computeTF(numOfWordsA, bagOfWordsA)
tfB = computeTF(numOfWordsB, bagOfWordsB)

# IDF
def computeIDF(documents):
    N = len(documents)
    idfDict = dict.fromkeys(documents[0].keys(), 0)

    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1

    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))

    return idfDict

idfs = computeIDF([numOfWordsA, numOfWordsB])

# TF-IDF
def computeTFIDF(tfBag, idfs):
    tfidf = {}
    for word, val in tfBag.items():
        tfidf[word] = val * idfs[word]
    return tfidf

tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)

df_manual = pd.DataFrame([tfidfA, tfidfB])
print(df_manual)

# ==========================================
# STEP 8: TF-IDF USING SKLEARN
# ==========================================
print("\n--- TF-IDF using sklearn ---")

from sklearn.feature_extraction.text import TfidfVectorizer

documents = [documentA, documentB]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

df_sklearn = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(df_sklearn)

# ==========================================
# DONE ✅
# ==========================================
print("\n--- Execution Completed Successfully ---")
```
