```python
for sp in df["species"].unique():
    print(f"\nStatistics for {sp}")
    print(df[df["species"]==sp].describe())
```

```python
grouped = df.groupby("species")
print(grouped.mean())
print(grouped.median())
print(grouped.std())
print(grouped.min())
print(grouped.max())
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    x="sepal_length",
    y="petal_length",
    hue="species",
    data=df
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x="petal_length", hue="species", fill=True)
plt.title("Density Distribution of Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x="species", y="petal_width", data=df, jitter=True)
plt.title("Strip Plot of Petal Width")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="species", y="petal_width", data=df)
plt.title("Petal Width Distribution")
plt.show()
```
