```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df.info()
df.describe()
```

```python
sns.histplot(df['age'], bins=10, kde=True)
plt.title("Distribution of Age")
plt.show()
```

```python
sns.jointplot(x='age', y='fare', data=df, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='age', y='fare', data=df, kind='hex')
plt.show()
```

```python
sns.rugplot(df['fare'])
plt.title("Rug Plot of Fare")
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=df)
plt.title("Average Age by Gender")
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=df, estimator=np.std)
plt.title("Std Deviation of Age by Gender")
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title("Passenger Count by Gender")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=df)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=df)
plt.title("Age Distribution by Gender & Survival")
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=df)
plt.title("Violin Plot of Age by Gender")
plt.show()
```

```python
sns.violinplot(x='sex', y='age', hue='survived', data=df)
plt.title("Violin Plot with Survival")
plt.show()
```

```python
sns.stripplot(x='sex', y='age', data=df, jitter=True)
plt.title("Strip Plot of Age")
plt.show()
```

```python
sns.stripplot(x='sex', y='age', hue='survived', data=df, jitter=True)
plt.title("Strip Plot with Survival")
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', data=df)
plt.title("Swarm Plot of Age")
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', hue='survived', data=df)
plt.title("Swarm Plot with Survival")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Annotated Correlation Heatmap")
plt.show()
```

```python
sns.clustermap(corr, cmap="viridis")
plt.show()
```
