```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import squarify
import numpy as np
from wordcloud import WordCloud
import plotly.graph_objects as go
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
plt.figure()
df['sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart - Gender")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
df['class'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart - Passenger Class")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
sizes=df['class'].value_counts()
labels=df['class'].value_counts().index
squarify.plot(sizes=sizes,label=labels,alpha=0.8)
plt.axis('off')
plt.title("Treemap - Passenger Class")
plt.show()
```

```python
plt.figure()
corr=df[['survived','pclass','age','sibsp','parch','fare']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Heatmap - Correlation")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df,x='age',y='fare')
plt.title("Scatter Plot - Age vs Fare")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df,x='age',y='fare',hue='survived')
plt.title("Scatter Plot - Survival")
plt.show()
```

```python
plt.figure()
sns.histplot(df['age'].dropna(),bins=20)
plt.title("Histogram - Age Distribution")
plt.show()
```

```python
plt.figure()
sns.histplot(df['fare'],bins=20)
plt.title("Histogram - Fare Distribution")
plt.show()
```

```python
text=" ".join(df['class'].astype(str))
wc=WordCloud(width=800,height=400,background_color='white').generate(text)
plt.figure()
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='class',y='fare',data=df)
plt.title("Box Plot - Fare by Class")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='sex',y='age',data=df)
plt.title("Box Plot - Age by Gender")
plt.show()
```

```python
plt.figure()
sns.barplot(x=df['class'].value_counts().index,y=df['class'].value_counts().values)
plt.title("Bar Chart - Passenger Class Count")
plt.show()
```

```python
plt.figure()
sns.barplot(x=df['sex'].value_counts().index,y=df['sex'].value_counts().values)
plt.title("Bar Chart - Gender Count")
plt.show()
```

```python
plt.figure()
plt.scatter(df['age'],df['fare'],s=df['fare'],alpha=0.5)
plt.title("Bubble Chart - Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
pivot=df.pivot_table(values='fare',index='class',columns='sex',aggfunc='mean')
pivot.plot(kind='bar',stacked=True)
plt.title("Stacked Column Chart")
plt.show()
```

```python
survived = df['survived'].value_counts()

labels = ["Start Passengers","Did Not Survive","Survived","Total"]
values = [0, survived[0], survived[1], 0]

fig = go.Figure(go.Waterfall(
    x = labels,
    y = values,
    measure = ["absolute","relative","relative","total"]
))

fig.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)+5
y2=np.cos(x)+5
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_trisurf(x,y1,y2,cmap='viridis')
plt.show()
```
